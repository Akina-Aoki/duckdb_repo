# DuckDB Python API
- In this lecture, instead of interacting with DuckDB through the terminal we are using Python Script.
This is why we are using the Duckdb Python API
- This is the same way of loading sakila data as via terminal.

### Connect DuckDB with sakila database through the python API

```python
import duckdb
```

### read data frames one by one

```python
# Use context manager to ensure the connection is properly closed.
# Connect to duckdb file with context manager

# duckdb path
duckdb_path = 'sakila.duckdb'

# duckdb function to connect to duckdb file and execute sql script to ingest sakila data
# this is the same as `duckdb sakila.duckdb < sql/load_sakila.sql` in terminal
with duckdb.connect(duckdb_path) as conn, open('sql/load_sakila.sql') as ingest_scripts:
    conn.sql(ingest_scripts.read())    # ingest sakila data

    # create pandas dataframe called `description` from the created duckdb file connection
    description = conn.sql("DESC;").df()     # get description of the database tables
    film = conn.sql("FROM film;").df()              # get all records from film table
```
